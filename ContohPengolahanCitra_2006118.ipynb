```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
from google.colab.patches import cv2_imshow  
img = cv2.imread('/content/drives/MyDrive/Colab Notebooks/fotbar.jpg', cv2.IMREAD_UNCHANGED)
cv2_imshow(img)
grayImg = cv2.cvtColor(np.array(img),cv2.COLOR_BGR2GRAY);
cv2_imshow(grayImg)
```

```python
plt.subplot(121), plt.imshow(img), plt.title("Sebelum di Edit")
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(grayImg), plt.title("Setelah di Edit")
plt.xticks([]), plt.yticks([])
plt.show()
```
