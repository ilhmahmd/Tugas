```python
import cv2 as cv
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
```

Grayscale

```python
gray_image = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

Histogram Grayscale

```python
plt.hist(gray_image.ravel(), bins = 256, range = [0,256])
plt.show()
```
